# 1. Tokenization Using Textblob

```python
from textblob import TextBlob

text = "TextBlob is very amazing and simple to use. What a great tool!"
blob = TextBlob(text)

print("Words:", blob.words)
print("Sentences:", blob.sentences)
```

# 2. Part of Speech (POS) Tagging

```python
print("POS Tags:", blob.tags)
```

| Tag  | Meaning                    | Example     |

# 3. Sentiment Analysis

```python
TextBlob("I love this phone.").sentiment

# polarity means whether the text is positive, negative, or neutral .
# Subjectivity tells whether the sentence is based on personal opinion or facts.
```

# Polarity
| Polarity Value | Meaning             |
| -------------- | ------------------- |
| -1.0           | Completely Negative |
| -0.5           | Negative            |
| 0.0            | Neutral             |
| +0.5           | Positive            |
| +1.0           | Completely Positive |


# Subjectivity
| Subjectivity Value | Meaning                     |
| ------------------ | --------------------------- |
| 0.0                | Completely Objective (Fact) |
| 0.3                | Mostly Fact                 |
| 0.5                | Mix of Fact and Opinion     |
| 0.8                | Mostly Opinion              |
| 1.0                | Completely Opinion          |


```python
review = "This mobile phone is amazing. The camera quality is excellent and the battery lasts all day. I love it!"

blob = TextBlob(review)

print("Review:")
print(review)

print("\nSentiment:")
print(blob.sentiment)
```
