# a. Parts of Speech (POS) tagging.

# What is POS in NLTK?

POS stands for Part of Speech.

It is the process of assigning a grammatical category to each word in a sentence, such as noun, verb, adjective, or adverb.

In NLTK (Natural Language Toolkit), POS tagging is performed using the pos_tag() function.

# Common POS Tags in NLTK
| POS Tag  | Full Form                             | Meaning                   | Example     |
| -------- | ------------------------------------- | ------------------------- | ----------- |
| **NN**   | Noun, Singular                        | Person, place, thing      | dog, book   |
| **NNS**  | Noun, Plural                          | More than one noun        | dogs, books |
| **NNP**  | Proper Noun                           | Specific name             | India, Om   |
| **NNPS** | Proper Noun Plural                    | Plural proper noun        | Americans   |
| **VB**   | Verb (Base Form)                      | Base verb                 | run, eat    |
| **VBD**  | Verb (Past Tense)                     | Past action               | played      |
| **VBG**  | Verb (Gerund)                         | Verb ending with *-ing*   | playing     |
| **VBN**  | Past Participle                       | Completed action          | written     |
| **VBP**  | Verb (Present)                        | Present tense             | play        |
| **VBZ**  | Verb (3rd Person Singular)            | Present singular          | plays       |
| **JJ**   | Adjective                             | Describes a noun          | beautiful   |
| **JJR**  | Comparative Adjective                 | Comparison                | bigger      |
| **JJS**  | Superlative Adjective                 | Highest degree            | biggest     |
| **RB**   | Adverb                                | Modifies a verb/adjective | quickly     |
| **RBR**  | Comparative Adverb                    | Faster comparison         | faster      |
| **RBS**  | Superlative Adverb                    | Fastest                   | fastest     |
| **DT**   | Determiner                            | Articles                  | the, a, an  |
| **PRP**  | Personal Pronoun                      | Pronouns                  | I, he, she  |
| **IN**   | Preposition/Subordinating Conjunction | in, on, at, because       | in, on      |
| **CC**   | Coordinating Conjunction              | Connects words            | and, or     |
| **CD**   | Cardinal Number                       | Numbers                   | 10, 100     |
| **UH**   | Interjection                          | Expression                | wow!, oh!   |


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize

stop_words=set(stopwords.words("english"))

txt = ("Artificial Intelligence is transforming the way people live and work. "
    "It enables computers to learn from data and make intelligent decisions. "
    "Machine Learning and Deep Learning are important branches of AI. "
    "Natural Language Processing helps computers understand human language. "
    "AI is widely used in healthcare, education, finance, and transportation. "
    "Researchers continue to develop advanced models that improve accuracy and efficiency. "
    "Learning AI requires a strong foundation in programming, mathematics, and data analysis.")

tokenized=sent_tokenize(txt)

for sentence in tokenized:
    words=word_tokenize(sentence)
    words=[w for w in words if w.lower() not in stop_words]
    tagged=nltk.pos_tag(words)
    print(tagged)
```

# b. Lemmatization to convert words into their meaningful base forms.
# What is Lemmatization with NLTK
Lemmatization is a text preprocessing technique in Natural Language Processing (NLP) that converts words into their base or dictionary form called a lemma. Unlike stemming, it considers the meaning and part of speech of words, making the output more accurate and meaningful.

```python
import nltk
nltk.download('punkt_tab')      
nltk.download('wordnet')    
nltk.download('omw-1.4') 
nltk.download('averaged_perceptron_tagger_eng')
```


| NLTK Resource                    | Purpose                                                                                                     |
| -------------------------------- | ----------------------------------------------------------------------------------------------------------- |
| `punkt_tab`                      | Provides tokenization data to split text into sentences and words.                                          |
| `wordnet`                        | Downloads the WordNet lexical database used for lemmatization.                                              |
| `omw-1.4`                        | Downloads Open Multilingual WordNet data to improve lemmatization support.                                  |
| `averaged_perceptron_tagger_eng` | Downloads the English POS tagging model used by `nltk.pos_tag()` to identify the grammatical role of words. |


```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
text = "The cats were running faster than the dogs."
tokens = word_tokenize(text)
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]

print(f"Original Text: {text}")
print(f"Lemmatized Words: {lemmatized_words}")
```
